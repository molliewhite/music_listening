```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-agent': 'Custom User Agent'}
```

```python
url = 'https://www.riaa.com/gold-platinum/?tab_active=awards_by_album#search_section'
```

```python
response = requests.get(url, headers=headers)
```

```python
table = pd.read_html(response.text)
table
```

```python
no_details = table[0]
no_details.head()
```

```python
no_details.fillna(value='All', inplace=True)
no_details.head()
```

```python
no_details.columns = no_details.iloc[0]
no_details.head()
```

```python
no_details_cleanup = no_details.drop(no_details.index[0])
no_details_cleanup.head()
```

```python
no_details_cleanup = no_details_cleanup[['Artist', 'Title', 'Label', 'Format', 'Certified Units (In Millions)']]
no_details_cleanup.head()
```

```python
c_units = no_details['Certified Units (In Millions)'].str.split(" ", n = 1, expand = True)
c_units.columns = c_units.iloc[0]
c_units = c_units.drop(c_units.index[0])
c_units.head()
```

```python
cu = c_units['Certified']
```

```python
no_details_cleanup['Certified Units (Millions)'] = cu
no_details_cleanup.head()
```

```python
no_md = no_details_cleanup.drop(no_details_cleanup.columns[4], axis=1)
no_md.head()
```

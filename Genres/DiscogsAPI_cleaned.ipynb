```python
#create empty list to hold respose data
genre_list = []

#perform multiple searches of the API database for genres based on artists, by iterrating through the artist column in scrapped df
for i, row in df.iterrows():
    if i > 200:
        break
    results = d.search(row['Artist'], type='release')
    time.sleep(2)
    try:
        genres = (results[0].genres, i)
        #print(genres)
        genre_list.append(genres)
    except:
        genre_list.append('unknown', i)
    
```

```python
#discogs requires that the requests not occur too quickly, but that the total elapsed time of the request also not 
#exceed a certain amount. Breaking the request into multiple parts, 200 requests each.
for i, row in df.iterrows():
    if i < 201:
        continue
    elif i > 400:
        break
    results = d.search(row['Artist'], type='release')
    time.sleep(2)
    try:
        genres = (results[0].genres, i)
        #print(genres)
        genre_list.append(genres)
    except:
        genre_list.append('unknown', i)
```
